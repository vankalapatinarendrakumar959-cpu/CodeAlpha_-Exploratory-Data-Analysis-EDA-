EDA Questions:

1. Which product has the highest sales?
2. Which city generates maximum revenue?
3. Does discount affect sales?
4. What is the average customer age?
5. Are there any outliers in sales?
6. Is there a relationship between rating and sales?

```python
sns.barplot(x="Product", y="Sales", data=df)

plt.title("Sales by Product")
plt.xticks(rotation=45)
plt.show()
```

```python
product_sales = df.groupby("Product")["Sales"].sum()

product_sales
```

```python
plt.figure(figsize=(8,5))

product_sales.plot(kind="bar")

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")

plt.show()
```

```python
city_sales = df.groupby("City")["Sales"].sum()

city_sales
```

```python
plt.figure(figsize=(8,5))

city_sales.plot(kind="bar")

plt.title("Revenue by City")
plt.xlabel("City")
plt.ylabel("Revenue")

plt.show()
```

```python
sns.scatterplot(
    x="Discount",
    y="Sales",
    data=df
)

plt.title("Discount vs Sales")

plt.show()
```

```python
average_age = df["Age"].mean()

average_age
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    y=df["Sales"]
)

plt.title("Sales Outlier Detection")

plt.show()
```

```python
sns.scatterplot(
    x="Rating",
    y="Sales",
    data=df
)

plt.title("Rating vs Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```
